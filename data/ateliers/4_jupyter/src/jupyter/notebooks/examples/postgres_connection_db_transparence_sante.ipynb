Ce notebook suppose que vous avez le postgres du projet transparence-sante up et que les données ont déjà été préprocessées et uploadées dans la db.

Il suffit de créer une connection :

```python
import psycopg2
conn = psycopg2.connect(
    host="postgres",
    database="postgres",
    user="postgres",
    password="")
```

De récupérer un curseur :

```python
c = conn.cursor()
```

D'éxécuter une requête :

```python
sql_query = 'SELECT date, nom_prénom FROM declaration LIMIT 10'
c.execute(sql_query)
```

De récupérer les résultats :

```python
results = c.fetchall()
for result in results:
    print(result)
```

Ou d'utiliser pandas :

```python
import pandas as pd

df = pd.read_sql('SELECT * FROM declaration LIMIT 10', conn)
df.head(10)
```
